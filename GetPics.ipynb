```python
import os 

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# defining global variable path
image_path = r"C:\Users\User\Documents\Dataset eye\Dataset\Dataset\Irisimage"


def loadImages(path):
    # Put files into lists and return them as one list of size 4
    image_files = [os.path.join(path, '1', file)
                     for file in os.listdir(path + "/1") if file.endswith('.JPG')]
    annotation_files = sorted([os.path.join(image_path, '2', file)
                     for file in os.listdir(path + "/2") if file.endswith('.JPG')])
    image_val_files = sorted([os.path.join(path, '3', file) 
                     for file in os.listdir(path + "/3") if file.endswith('.JPG')])
    annotation_val_files = sorted([os.path.join(path, "4", file) 
                     for file in os.listdir(path + "/4") if file.endswith('.JPG')])
    return image_files, annotation_files, image_val_files, annotation_val_files

def main():
    
    # calling global variable
    global image_path
    
    '''The var Dataset is a list of size 4 that gets all files from   
    different folders so we can access different folders using dataset[0...3] '''
    dataset = loadImages(image_path)
    train_set = dataset[0]

    r = [i for i in train_set]
    print(r[4]) 
        
    img_1 = mpimg.imread(r[4])
    display = plt.imshow(img_1)
    print(img_1)
    
main()
```
